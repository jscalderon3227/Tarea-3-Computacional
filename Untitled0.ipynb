```python


x=zeros(1000)
y=zeros(1000)
z=zeros(1000)
```

```python
u=random.random(1000)
v=random.random(1000)
```

```python
theta=2*pi*u
phi=arccos(2*v-ones(1000))
```

```python
x=sin(theta)*cos(phi)
y=sin(theta)*sin(phi)
z=cos(theta)
```

```python
scatter(x,y)
```

```python
scatter(x,z)
```

```python
fig = pylab.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z)
```
