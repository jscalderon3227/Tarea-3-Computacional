```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
u=random.random(1000)
r=u**(1.0/(3.0))
```

```python
v=random.random(1000)
theta=2.0*pi*v

w=random.random(1000)
phi=arccos((2.0*w)-1.0)
```

```python
x=r*sin(theta)*cos(phi)
y=r*sin(theta)*sin(phi)
z=r*cos(theta)
```

```python
scatter(x,y)
```

```python
scatter(x,z)
```

```python
fig = pylab.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z)
```
