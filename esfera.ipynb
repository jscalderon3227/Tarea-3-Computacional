```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=2*random.random(2000)-1
y=2*random.random(2000)-1
z=2*random.random(2000)-1
r=sqrt(((x)**2)+((y)**2)+((z)**2))

```

```python
index=where(r<1)
```

```python
n=size(index)

```

```python
lista=index[0]
n=size(lista)
ejex=zeros(n)
ejey=zeros(n)
ejez=zeros(n)

for i in range(n):
    ejex[i]=x[lista[i]]
    ejey[i]=y[lista[i]]
    ejez[i]=z[lista[i]]
    
```

```python
scatter(ejex,ejey)
```

```python
scatter(ejex,ejez)
```

```python
fig = pylab.figure()
ax = Axes3D(fig)
ax.scatter(ejex,ejey,ejez)
```
